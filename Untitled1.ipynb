```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "Salary Data.csv"
df = pd.read_csv(file_path)
```

```python
# 3. Basic Data Inspection
print("Shape of dataset:", df.shape)
print("\nColumn Names:\n", df.columns)
print("\nDataset Info:")
print(df.info())
print("\nDuplicated: \n",df.apply(lambda x: x.duplicated().sum()))
```

```python
# 4. Missing Values Analysis
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Percentage of missing values
missing_percent = (df.isnull().sum() / len(df)) * 100
print("\nMissing Values Percentage:")
print(missing_percent)
```

```python
# 5. Statistical Summary
print("\nStatistical Summary (Numerical):")
print(df.describe())

print("\nStatistical Summary (Categorical):")
print(df.describe(include="object"))
```

```python
# 6. Univariate Analysis (Numerical)
numerical_cols = ["Age", "Years of Experience", "Salary"]

for col in numerical_cols:
    plt.figure()
    sns.histplot(df[col], bins=20, kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# 7. Univariate Analysis (Categorical)
categorical_cols = ["Gender", "Education Level", "Job Title"]
for col in categorical_cols:
    plt.figure()
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f"Count Plot of {col}")
    plt.xlabel("Count")
    plt.ylabel(col)
    plt.show()
```

```python
# 8. Bivariate Analysis
# =========================================

# Salary vs Years of Experience
plt.figure()
sns.scatterplot(x="Years of Experience", y="Salary", data=df)
plt.title("Salary vs Years of Experience")
plt.show()

# Salary vs Education Level
plt.figure()
sns.boxplot
```

```python
# Salary vs Years of Experience
plt.figure()
sns.scatterplot(x="Years of Experience", y="Salary", data=df)
plt.title("Salary vs Years of Experience")
plt.show()

# Salary vs Education Level
plt.figure()
sns.boxplot(x="Education Level", y="Salary", data=df)
plt.title("Salary by Education Level")
plt.show()

# Salary vs Gender
plt.figure()
sns.boxplot(x="Gender", y="Salary", data=df)
plt.title("Salary by Gender")
plt.show()
```

```python
# 9. Correlation Analysis
# =========================================
plt.figure()
corr = df[numerical_cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 11. Outlier Detection (IQR Method)
# =========================================
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"\nOutliers detected in {col}: {outliers.shape[0]}")
```

```python
# Example: Drop missing values
df_cleaned = df.dropna()

print("\nShape after removing missing values:", df_cleaned.shape)
print("\nEDA Completed Successfully!")
```
